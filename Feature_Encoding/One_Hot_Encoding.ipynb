# 🧠 Important Concepts

## 📌 What is One Hot Encoding?

One Hot Encoding is a feature encoding technique that converts **categorical data** into **binary numerical vectors**.

It creates a separate column for each unique category and represents the presence of a category with **1** and its absence with **0**.

---

## ⚙️ How to Implement One Hot Encoding?

### Using Pandas

```python
pd.get_dummies(data, columns=["Column_Name"], dtype=int)
```

### Using Scikit-learn

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(data[["Column_Name"]])
```

---

## 📌 Key Points

- One column is created for each **unique category**.
- The **number of rows remains the same**.
- Best suited for **Nominal Categorical Data**.
- Not recommended for **high-cardinality** features because it creates many columns and sparse matrices due to which memory wasted .

````python
import pandas as pd
````

````python
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red", "Blue"]
})
df
````

````python
print("unique categories:")
print(df["Color"].unique())
````

````python
print(df["Color"].value_counts())
````

````python
encoded_DF =pd.get_dummies(df,columns=["Color"] )
print(encoded_DF)
````

````python
encoded_DF =pd.get_dummies(df,columns=["Color"] , dtype=int )
print(encoded_DF)
````

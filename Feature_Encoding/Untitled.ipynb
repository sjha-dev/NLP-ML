# 🏷️ Feature Encoding

Welcome to the **Feature Encoding** section of the **NLP-ML** repository.

Feature Encoding is the process of converting categorical data into a numerical format that Machine Learning algorithms can understand.

Since most Machine Learning models cannot work directly with text or categorical values, feature encoding is an essential preprocessing step.

# 📖 One Hot Encoding

One Hot Encoding is one of the most commonly used encoding techniques for **nominal categorical data**.

Instead of assigning numbers to categories, it creates a separate binary column for each unique category.

### Example

Original Data

| Color |
|--------|
| Red |
| Blue |
| Green |

After One Hot Encoding

| Color_Red | Color_Blue | Color_Green |
|------------|------------|-------------|
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 0 | 0 | 1 |

Each category is represented by:

- **1** → Category is present
- **0** → Category is absent

---

# 🎯 Why Do We Need Feature Encoding?

Machine Learning algorithms work with numbers, not text.

For example:

```text
Color
Red
Blue
Green
```

cannot be directly used for training a model.

Feature Encoding converts such categorical values into numerical representations while preserving useful information.

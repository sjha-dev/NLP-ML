# Named Entity Recognition (NER)

## What is Named Entity Recognition?

Named Entity Recognition (NER) is a Natural Language Processing (NLP) technique used to identify and classify important entities in a text into predefined categories.

Common entity categories include:

- Person Names
- Organizations
- Locations
- Dates
- Times
- Money
- Percentages

NER helps machines understand the real-world meaning of words and phrases in a sentence.

---

## Example 1

Sentence:

```text
Sudhanshu lives in Delhi and works at Google.
```

NER Output:

| Entity | Entity Type |
|----------|------------|
| Sudhanshu | PERSON |
| Delhi | GPE (Location) |
| Google | ORGANIZATION |

---

## Example 2

Sentence:

```text

```

NER Output:

| Entity | Entity Type |
|----------|------------|
| Elon Musk | PERSON |
| Tesla | ORGANIZATION |

---

## Common Entity Types

| Entity Type | Description |
|------------|-------------|
| PERSON | Names of people |
| ORGANIZATION (ORG) | Companies, institutions, organizations |
| GPE | Countries, cities, states |
| LOCATION | Physical locations |
| DATE | Dates |
| TIME | Time expressions |
| MONEY | Monetary values |
| PERCENT | Percentages |

---

## POS Tagging vs NER

### POS Tagging

POS Tagging identifies the grammatical role of a word.

Example:

```text
Google → NNP (Proper Noun)
Delhi → NNP (Proper Noun)
Sudhanshu → NNP (Proper Noun)
```

### Named Entity Recognition

NER identifies the actual meaning of the entity.

Example:

```text
Google → ORGANIZATION
Delhi → LOCATION
Sudhanshu → PERSON
```

Thus, NER provides more meaningful information than POS Tagging.

---

## Why is NER Important?

- Information Extraction
- Search Engines
- Chatbots and Virtual Assistants
- Question Answering Systems
- News Analysis
- Recommendation Systems

NER helps computers understand important information from unstructured text.

---

## NLTK Implementation

```python
import nltk
from nltk.tokenize import word_tokenize

sentence = "Sudhanshu lives in Delhi and works at Google"

words = word_tokenize(sentence)

pos_tags = nltk.pos_tag(words)

named_entities = nltk.ne_chunk(pos_tags)

print(named_entities)
```

## NLP Pipeline

```text
Raw Text
    ↓
Sentence Tokenization
    ↓
Word Tokenization
    ↓
Stopwords Removal
    ↓
Stemming / Lemmatization
    ↓
POS Tagging
    ↓
Named Entity Recognition (NER)
    ↓
Feature Extraction
    ↓
Machine Learning / Deep Learning
```

### Q: Why do we remove stop words?

### Answer:
### Stop words are very common words such as is, am, are, the, and. They usually do not contribute much meaning to the text. Removing them reduces noise, decreases dimensionality, and improves NLP model efficiency.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords 
```

```python
stopwords.words('english')                   ## give all stopwords used in english..
```

```python
paragraph = '''All Birds find shelter during a rain.
But Eagle avoids rain by flying above
the Clouds. Problems are common, but attitude
makes the difference!! '''

stop_words=set(stopwords.words('english'))
words=paragraph.split()
filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)
print(filtered_words)        

```

### What is the NLP pipeline ? 

### A pipeline is a sequence of processing steps where the output of one step becomes the input of the next step. In NLP, a text preprocessing pipeline may include tokenization, stop word removal, lemmatization, vectorization, and model training.

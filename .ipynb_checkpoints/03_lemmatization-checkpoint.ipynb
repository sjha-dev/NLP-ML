## lemmatization 
### -------------------------------------------------------------------------------------------------

### Lemmatization is the process of converting a word to its base or dictionary form (called a lemma) by considering the word's meaning and grammatical context.
### Unlike stemming, lemmatization produces meaningful words

### examples : 
### =================================================================================================
### | Word    | Lemmatization Output |
### | ------- | -------------------- |
### | running | run                  |
### | studies | study                |
### | better  | good                 |
### | went    | go                   |
### | cars    | car                  |

### How it works ? 
### answer :- Lemmatization uses:

### Vocabulary (dictionary)
### Morphological analysis
### Part-of-Speech (POS) information ( verb - v , noun - n , adverb - a , )


```python
import nltk 
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
wnk = WordNetLemmatizer()
```

```python
print(wnk.lemmatize("crying"))
```

```python
print(wnk.lemmatize("running"))
```

```python
print(wnk.lemmatize("running",pos='v'))
```

```python
print(wnk.lemmatize("better",pos='a'))
```

```python
print(wnk.lemmatize("excellent",pos='a'))
```

```python
import nltk 
nltk.download('punkt_tab')
nltk.download('punkt')
```

```python
corpus=""" Hey , my name is sudhanshu jha. 
            what are your name ? 
            tell me about yourself.
            please watch every video on my yt.
       """
```

```python
## tokenization 
## paragraph to sentences 

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## tokenization
## paragraph --> words 
## sentences --> words 

from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```

```python
print(words)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

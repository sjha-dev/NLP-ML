# Part of Speech (POS) Tagging

## What is POS Tagging?

Part of Speech (POS) Tagging is the process of assigning a grammatical category to each word in a sentence.

These categories include:

- Noun (NN)
- Verb (VB)
- Adjective (JJ)
- Adverb (RB)
- Pronoun (PRP)
- Proper Noun (NNP)

POS Tagging helps computers understand the role of each word in a sentence.

---

## Example 1

Sentence:

```text
The beautiful girl sings beautifully.
```

POS Tags:

| Word | POS Tag | Meaning |
|--------|----------|----------|
| The | DT | Determiner |
| beautiful | JJ | Adjective |
| girl | NN | Noun |
| sings | VBZ | Verb |
| beautifully | RB | Adverb |

---

## Example 2

Sentence:

```text
Sudhanshu is learning NLP.
```

POS Tags:

| Word | POS Tag | Meaning |
|--------|----------|----------|
| Sudhanshu | NNP | Proper Noun |
| is | VBZ | Verb |
| learning | VBG | Verb (Present Participle) |
| NLP | NNP | Proper Noun |

---

## Common POS Tags

| Tag | Description |
|------|-------------|
| NN | Noun |
| NNS | Plural Noun |
| NNP | Proper Noun |
| JJ | Adjective |
| RB | Adverb |
| VB | Verb |
| VBD | Past Tense Verb |
| VBG | Verb + ing Form |
| PRP | Pronoun |
| DT | Determiner |

---

## Why is POS Tagging Important?

- Helps understand sentence structure.
- Improves text analysis.
- Used in chatbots and virtual assistants.
- Useful for Named Entity Recognition (NER).
- Helps in machine translation and sentiment analysis.

---

## NLTK Example

```python
import nltk
from nltk.tokenize import word_tokenize

sentence = "Sudhanshu is learning NLP"

words = word_tokenize(sentence)

pos_tags = nltk.pos_tag(words)

print(pos_tags)
```

Output:

```python
[
 ('Sudhanshu', 'NNP'),
 ('is', 'VBZ'),
 ('learning', 'VBG'),
 ('NLP', 'NNP')
]
```

---

## Conclusion

POS Tagging is an important NLP technique that assigns grammatical labels to words and helps machines understand the meaning and structure of sentences.

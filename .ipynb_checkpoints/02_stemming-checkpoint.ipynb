## stemming --> Stemming is a Natural Language Processing (NLP) technique used to reduce inflected or derived words to their stem (root) form by removing affixes such as prefixes and suffixes.

### example : "connect", "connected", "connecting", and "connection" → connect

### Purpose:
###          Reduce vocabulary size
###          Improve text processing efficiency
###          Help in information retrieval and text mining

## there are few famous algorithms of stemming 

## 1) PorterStemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
words = ["playing","played","studies","connection","connecting","connected","lively","living","lived"]
```

```python
for word in words:
    print(word+ "->"+ ps.stem(word))
```

```python
ps.stem("congratulations")
```

## 2) RegexpStemmer class 

```python
from nltk.stem import RegexpStemmer
stemming = RegexpStemmer('ing$|ed$|s$', min = 4)
print(stemming.stem("playing"))
print(stemming.stem("studying"))
print(stemming.stem("singing"))
print(stemming.stem("sings"))
print(stemming.stem("played"))
print(stemming.stem("books"))
print(stemming.stem("shocks"))
```

## 3) Snowball Stemmer :- It is the improved version of porter stemmer but not too much improved



```python
from nltk.stem import SnowballStemmer
```

```python
ss = SnowballStemmer('english')
print(ss.stem("playing"))
```

```python
print(ss.stem("fairly")) , print(ss.stem("sparingly"))
```
